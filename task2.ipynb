```python
import pandas as pd
```

```python
df = pd.read_csv("Dry_Bean.csv")
df.head()
```

```python
corr_matrix = df.corr(numeric_only=True)
corr_matrix.head(4)
```

```python
# features (absolute value, regardless of the direction)
corrfeat = corr_matrix[abs(corr_matrix) > 0.8]
corrfeat = corrfeat.stack().reset_index()
corrfeat = corrfeat.rename(columns={'level_0': 'Feature 1', 'level_1': 'Feature 2', 0: 'Correlation'})
pd.set_option('display.max_rows', None)
corrfeat
#corrfeat.count() #78
```

```python
# features (only positive direction)
corrfeat = corr_matrix[corr_matrix > 0.8]
corrfeat = corrfeat.stack().reset_index()
corrfeat = corrfeat.rename(columns={'level_0': 'Feature 1', 'level_1': 'Feature 2', 0: 'Correlation'})
pd.set_option('display.max_rows', None)
corrfeat
#corrfeat.count() #54
```
