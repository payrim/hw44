```python
import pandas as pd
```

```python
df = pd.read_csv("Dry_Bean.csv")
df.head()
```

```python
# Mean
mean_values = df.mean(numeric_only=True)
df_mean = pd.DataFrame([mean_values], columns=df.columns)
df_mean
```

```python
# Standard deviation
std_values = df.std(numeric_only=True)
df_std = pd.DataFrame([std_values], columns=df.columns)
df_std
```

```python
# Minimum
min_values = df.min(numeric_only=True)
df_min = pd.DataFrame([min_values], columns=df.columns)
df_min
```

```python
# 25th percentile (Q1)
q1_values = df.quantile(0.25, numeric_only=True)
df_q1 = pd.DataFrame([q1_values], columns=df.columns)
df_q1
```

```python
# Median (50th percentile)
median_values = df.median(numeric_only=True)
df_median = pd.DataFrame([median_values], columns=df.columns)
df_median
```

```python
# 75th percentile (Q3)
q3_values = df.quantile(0.75, numeric_only=True)
df_q3 = pd.DataFrame([q3_values], columns=df.columns)
df_q3
```

```python
# Maximum
max_values = df.max(numeric_only=True)
df_max = pd.DataFrame([max_values], columns=df.columns)
df_max
```

```python
# Count
count_values = df.count()
df_count = pd.DataFrame([count_values], columns=df.columns)
df_count
```

```python
# Sum
sum_values = df.sum(numeric_only=True)
df_sum = pd.DataFrame([sum_values], columns=df.columns)
df_sum
```

```python
# Create a new DataFrame from the calculated values
summary_df = pd.DataFrame(
    [std_values, min_values, q1_values, median_values, q3_values, max_values, count_values, sum_values],
    index=['Standard Deviation', 'Minimum', '25th Percentile', 'Median', '75th Percentile', 'Maximum', 'Count', 'Sum']
)
summary_df
```
